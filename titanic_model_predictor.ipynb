```python
# Import important libraries 

import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
import_data = pd.read_csv("titanic_dataset.csv")
import_data.head()
```

```python
import_data.info()
```

```python
import_data.isnull().sum()
```

####  Age,Embarked and Cabin columns has null values.

```python
import_data[['Age']].describe().T # Age Column.
```

```python
unique_values_of_cabin = import_data['Cabin'].unique()
unique_values_of_cabin
```

```python
mean_of_age = import_data['Age'].mean()
mode_of_cabin = import_data['Cabin'].mode()
mode_of_embarked = import_data['Embarked'].mode()

not_null_df = import_data.copy()
not_null_df['Age'] = not_null_df['Age'].fillna(mean_of_age )
not_null_df['Cabin'] = not_null_df['Cabin'].fillna(mode_of_cabin[0])  
not_null_df['Embarked'] = not_null_df['Embarked'].fillna(mode_of_embarked[0])
```

```python
not_null_df.isnull().sum()
```

### So , now we have a clean dataset for titanic.

#### Let's Change categorical columns to Numerical.

```python
cleaned_data = not_null_df.drop(columns=['Name','PassengerId','Ticket','Cabin'])

cleaned_data.head()
```

```python
s = (cleaned_data.dtypes == 'object')
object_cols = s[s].index 
print(f"Categorical columns in cleaned data: {object_cols}")
```

```python
from sklearn.preprocessing import OneHotEncoder 
from sklearn.compose import make_column_transformer 

encoder = OneHotEncoder(categories = 'auto',
                        handle_unknown='ignore',
                        drop = 'first',
                        sparse= True
                        )


transformer = make_column_transformer(
                                    (
                                        OneHotEncoder(),
                                        ['Sex','Embarked']
                                    ),
                                    remainder = 'passthrough')

transformed = transformer.fit_transform(cleaned_data[object_cols])
transformed_df = pd.DataFrame(transformed,
                              columns = transformer.get_feature_names()
                             )

transformed_df.head()
```

```python
final_df = pd.concat([cleaned_data,transformed_df],axis = 1)
final_df.head()
```

```python
final_df = final_df.drop(['onehotencoder__x1_C','Sex','Embarked'],axis = 1)
```
